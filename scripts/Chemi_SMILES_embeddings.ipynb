```
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
model = AutoModel.from_pretrained("seyonec/ChemBERTa-zinc-base-v1")
model = model.eval()
```

```
import pandas as pd

df = pd.read_csv("/content/combined_compound_SMILES_BBclass_cleaned.csv")
smiles_list = df["canonical_smiles"].tolist()
```

```
import numpy as np
from tqdm import tqdm

embeddings = []

for smi in tqdm(smiles_list):
    inputs = tokenizer(smi, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    emb = outputs.last_hidden_state.mean(dim=1).detach().numpy()
    embeddings.append(emb)

embeddings = np.vstack(embeddings)
np.save("all_embeddings.npy", embeddings)
df.to_csv("embedding_order.csv", index=True)
```
